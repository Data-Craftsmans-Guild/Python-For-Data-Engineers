# Scraping Basic HTML Tables with Pandas
https://www.treasury.gov/resource-center/data-chart-center/interest-rates/Pages/TextView.aspx?data=yieldYear&year=2020

```python
import pandas as pd
```

```python
tables = pd.read_html("https://www.treasury.gov/resource-center/data-chart-center/interest-rates/Pages/TextView.aspx?data=yieldYear&year=2020")
tables[3]
```

```python
tables[3].to_csv("Treasury Rates Yield Curve.csv", index=False, header=True)
```
